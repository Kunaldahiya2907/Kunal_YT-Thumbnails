```python
import cv2
import os
import numpy as np
import dlib
import pandas as pd
from deepface import DeepFace
from PIL import Image, ImageEnhance

# Load face detector
detector = dlib.get_frontal_face_detector()

# Function to calculate brightness
def calculate_brightness(image):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    return np.mean(hsv[:, :, 2])  # V channel

# Function to calculate contrast
def calculate_contrast(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    return gray.std()  # Standard deviation of pixel intensities

# Function to calculate saturation
def calculate_saturation(image):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    return np.mean(hsv[:, :, 1])  # S channel

# Function to calculate hue
def calculate_hue(image):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    return np.mean(hsv[:, :, 0])  # H channel

# Function to calculate sharpness
def calculate_sharpness(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    laplacian_var = cv2.Laplacian(gray, cv2.CV_64F).var()
    return laplacian_var

# Function to estimate gamma correction
def estimate_gamma(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    gamma = np.log(np.mean(gray) / 255.0) / np.log(0.5)
    return gamma

# Process thumbnails
thumbnail_folder = "thumbnails"  # Adjust path as needed
results = []

if os.path.exists(thumbnail_folder):
    for filename in os.listdir(thumbnail_folder):
        if filename.endswith(".jpg") or filename.endswith(".png"):
            filepath = os.path.join(thumbnail_folder, filename)
            image = cv2.imread(filepath)

            # Get image properties
            height, width, _ = image.shape
            resolution = height * width
            aspect_ratio = width / height

            # Convert to PIL for better analysis
            pil_image = Image.open(filepath)
            enhancer = ImageEnhance.Color(pil_image)

            # Calculate white balance deviation
            r, g, b = np.mean(np.array(pil_image), axis=(0, 1))
            white_balance_deviation = np.std([r, g, b])

            # Face detection
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            faces = detector(gray)
            face_present = 1 if len(faces) > 0 else 0
            num_faces = len(faces)
            face_size_ratios = [face.area() / resolution for face in faces]

            # Emotion detection (try-except in case of failure)
            try:
                analysis = DeepFace.analyze(filepath, actions=['emotion'], enforce_detection=False)
                dominant_emotion = analysis[0]['dominant_emotion']
            except:
                dominant_emotion = "unknown"

            # Store results
            results.append({
                "video_id": filename.replace(".jpg", "").replace(".png", ""),
                "face_present": face_present,
                "num_faces": num_faces,
                "face_size_ratios": face_size_ratios if num_faces > 0 else None,
                "dominant_emotions": dominant_emotion,
                "width": width,
                "height": height,
                "resolution": resolution,
                "aspect_ratio": aspect_ratio,
                "brightness": calculate_brightness(image),
                "contrast": calculate_contrast(image),
                "saturation": calculate_saturation(image),
                "hue": calculate_hue(image),
                "estimated_gamma": estimate_gamma(image),
                "sharpness": calculate_sharpness(image),
                "white_balance_deviation": white_balance_deviation,
                "exposure": calculate_brightness(image)  # Using brightness as proxy
            })

# Save results to CSV
df = pd.DataFrame(results)
df.to_csv("thumbnail_analysis.csv", index=False)

print("Thumbnail analysis completed. Data saved to 'thumbnail_analysis.csv'.")
```

```python
# Install necessary libraries
!pip install pandas numpy scikit-learn matplotlib seaborn

# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_absolute_error, r2_score

# Load the dataset
file_path = "/content/merged_video_data.csv"  # Update path if necessary
df = pd.read_csv(file_path)

# Display dataset information
print("Dataset Overview:")
print(df.info())
print(df.describe())

# Ensure only numeric columns are used for mean calculations
numeric_columns = df.select_dtypes(include=[np.number]).columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

# Convert categorical variables (e.g., dominant emotions) to numerical
if 'dominant_emotions' in df.columns:
    df = pd.get_dummies(df, columns=['dominant_emotions'], drop_first=True)

# Convert face_size_ratios (if stored as string lists) into numeric mean values
if 'face_size_ratios' in df.columns:
    df['face_size_ratios'] = df['face_size_ratios'].apply(lambda x: np.mean(eval(x)) if isinstance(x, str) and x.startswith('[') else x)

# Define target variable (Y) and features (X)
target_column = "views"  # Dependent variable
features = ['num_faces', 'face_size_ratios', 'width', 'height', 'resolution',
            'aspect_ratio', 'brightness', 'contrast', 'saturation', 'hue',
            'estimated_gamma', 'sharpness', 'white_balance_deviation', 'likes', 'comments']

# Keep only existing numeric columns
features = [col for col in features if col in df.columns]

# Prepare the dataset
X = df[features]
Y = df[target_column]

# Handle missing values in X (independent variables)
imputer = SimpleImputer(strategy="mean")
X = pd.DataFrame(imputer.fit_transform(X), columns=features)

# Train-Test Split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Train Linear Regression Model
lr_model = LinearRegression()
lr_model.fit(X_train, Y_train)

# Train Random Forest Regressor
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, Y_train)

# Predictions
Y_pred_lr = lr_model.predict(X_test)
Y_pred_rf = rf_model.predict(X_test)

# Evaluate Models
print("\nRegression Analysis Results:")
print(f"Linear Regression R² Score: {r2_score(Y_test, Y_pred_lr):.4f}")
print(f"Random Forest R² Score: {r2_score(Y_test, Y_pred_rf):.4f}")

print(f"Linear Regression MAE: {mean_absolute_error(Y_test, Y_pred_lr):.4f}")
print(f"Random Forest MAE: {mean_absolute_error(Y_test, Y_pred_rf):.4f}")

# Plot Actual vs Predicted for Random Forest
plt.figure(figsize=(8, 6))
sns.scatterplot(x=Y_test, y=Y_pred_rf, alpha=0.7)
plt.xlabel("Actual Views")
plt.ylabel("Predicted Views")
plt.title("Actual vs Predicted Views (Random Forest)")
plt.show()

# Feature Importance (Random Forest)
importances = rf_model.feature_importances_
feature_importance_df = pd.DataFrame({"Feature": features, "Importance": importances})
feature_importance_df = feature_importance_df.sort_values(by="Importance", ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x="Importance", y="Feature", data=feature_importance_df, palette="viridis")
plt.title("Feature Importance in Views Prediction")
plt.show()

# Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df[features + [target_column]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Install necessary libraries
!pip install pandas numpy scikit-learn matplotlib seaborn scipy

# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

# Load the dataset
file_path = "/content/merged_video_data.csv"  # Update path if necessary
df = pd.read_csv(file_path)

# Ensure 'face_present' and 'views' columns exist
if "face_present" not in df.columns or "views" not in df.columns:
    raise KeyError("The dataset must contain 'face_present' and 'views' columns.")

# Convert face_present column to binary values if necessary
df['face_present'] = df['face_present'].astype(int)

# Separate videos with and without faces
views_with_faces = df[df['face_present'] == 1]['views']
views_without_faces = df[df['face_present'] == 0]['views']

# Perform t-test to check for statistical significance
t_stat, p_value = ttest_ind(views_with_faces, views_without_faces, equal_var=False)

# Print results
print("\nFace Presence vs Views Analysis:")
print(f"Mean Views (With Faces): {views_with_faces.mean():,.2f}")
print(f"Mean Views (Without Faces): {views_without_faces.mean():,.2f}")
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print("Result: There is a statistically significant difference in views between thumbnails with and without faces.")
else:
    print("Result: There is no statistically significant difference in views between thumbnails with and without faces.")

# Visualization
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['face_present'], y=df['views'], palette="coolwarm")
plt.xticks(ticks=[0, 1], labels=["No Face", "Face Present"])
plt.xlabel("Presence of Face")
plt.ylabel("View Count")
plt.title("Impact of Face Presence on Video Views")
plt.yscale("log")  # Log scale to handle large variations in view counts
plt.show()
```
